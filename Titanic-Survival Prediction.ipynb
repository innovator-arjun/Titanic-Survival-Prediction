```python
import numpy as np
```

```python
tit_train['Age']=tit_train['Age'].fillna(-0.5)
tit_test['Age']=tit_test['Age'].fillna(-0.5)

bins=[-1,0,5,12,18,24,35,60,np.inf]

labels=['Unknown','Baby','Child','Teenager','Student','Young Adult','Adult','Senior']

tit_train['AgeGroup']=pd.cut(tit_train['Age'],bins,labels=labels)

tit_test['AgeGroup']=pd.cut(tit_test['Age'],bins,labels=labels)

```

```python
sns.barplot(x='Survived',y='AgeGroup',data=tit_train)
```
